```python
def scores_over_time(txt):
    jsn = json.loads(txt)
    rows = []
    for id in jsn['members']:
        member = jsn['members'][id]
        name = member['name']
        stars = member['completion_day_level']
        for s in stars:
            for part in stars[s]:
                star_epoch = stars[s][part]['get_star_ts']
                rows.append([id, name, s, part, dt.datetime.fromtimestamp(star_epoch)])
    df_raw = pd.DataFrame(rows, columns=['id','name','star','part','time'])
    df = df_raw.copy()
    n_members = df.id.nunique()
    df_new = [] 
    for i in (df.star + ',' + df.part).unique():
        star, part = i.split(',')
        df_star = df[(df.star == star) & (df.part == part)]
        df_star = df_star.sort_values(by=['time'], ascending=False)
        df_star['points_awarded'] = [j+1 for j in range(n_members-df_star.shape[0], n_members)]
        df_new.append(df_star)
    df = pd.concat(df_new)
    df.sort_values(['time'], inplace=True)
    

    df['local_score'] = 0
    for i in range(df.shape[0]):
        df_sub = df[(df.name == df.loc[i, 'name']) & (df.time <= df.loc[i, 'time'])]
        df.loc[i, 'local_score'] = df_sub.points_awarded.sum()
    df = df.sort_values(['time'])
    df = df[df.time < dt.datetime(df.time[0].year + 1, 1, 1)]

    df_current = df.groupby(['id']).time.max().reset_index()
    df_current = df_current.merge(df, on=['id', 'time'])
    df_current['time'] = df_current.time.max()
    df_out = pd.concat([df, df_current])
    
    
    return df_out.drop_duplicates().reset_index(drop=True)
```
